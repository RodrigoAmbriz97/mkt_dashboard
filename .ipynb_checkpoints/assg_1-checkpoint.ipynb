```python
import pandas as pd
import numpy as np
```

```python
fb = pd.read_csv('01_facebook_ads.csv') 
gl = pd.read_csv('02_google_ads.csv')
tt = pd.read_csv('03_tiktok_ads.csv')
```

```python
def campaign_summary(df, cost_col="cost"):
    """Group by campaign_name and compute key metrics."""
    agg = df.groupby(["campaign_name"], as_index=False).agg(
        total_conversions=("conversions", "sum"),
        total_cost=(cost_col, "sum"),
        total_clicks=("clicks", "sum"),
        total_impressions=("impressions", "sum"),
    )
    agg["clicks_to_conversion"]      = agg["total_clicks"] / agg["total_conversions"]
    agg["impressions_to_conversion"] = agg["total_impressions"] / agg["total_conversions"]
    agg["cost_per_click"]            = agg["total_cost"] / agg["total_clicks"]
    return agg.round(2)

fb_summary = campaign_summary(fb, cost_col="spend")
gl_summary = campaign_summary(gl)
tt_summary = campaign_summary(tt)

print("=== Facebook Ads ===")
display(fb_summary)
print("\n=== Google Ads ===")
display(gl_summary)
print("\n=== TikTok Ads ===")
display(tt_summary)
```

```python
def platform_summary(df, platform, cost_col="cost"):
    """Aggregate the entire table into a single row of KPIs."""
    total_conversions = df["conversions"].sum()
    total_cost        = df[cost_col].sum()
    total_clicks      = df["clicks"].sum()
    total_impressions = df["impressions"].sum()
    return pd.DataFrame([{
        "platform":                  platform,
        "total_conversions":         total_conversions,
        "total_cost":                round(total_cost, 2),
        "total_clicks":              total_clicks,
        "total_impressions":         total_impressions,
        "clicks_to_conversion":      round(total_clicks / total_conversions, 2),
        "impressions_to_conversion": round(total_impressions / total_conversions, 2),
        "cost_per_conversion":            round(total_cost / total_conversions, 2)
    }])

kpi = pd.concat([
    platform_summary(fb, "Facebook", cost_col="spend"),
    platform_summary(gl, "Google"),
    platform_summary(tt, "TikTok"),
], ignore_index=True)

display(kpi)
```
